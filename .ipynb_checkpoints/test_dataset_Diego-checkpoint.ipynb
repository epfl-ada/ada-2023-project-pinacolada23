```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import networkx as nx
import statsmodels.api as sm
import statsmodels.formula.api as smf

from scipy import stats
```

```python
# Load data
DATA_FOLDER = './MovieSummaries'
CHARACTER_DATASET = DATA_FOLDER + '\character.metadata.tsv'
MOVIE_DATASET = DATA_FOLDER + '\movie.metadata.tsv'
NAME_DATASET = DATA_FOLDER + '\name.clusters.txt'
PLOT_DATASET = DATA_FOLDER + '\plot_summaries.txt'
TVTROPES_DATASET = DATA_FOLDER + '\tvtropes.clusters.txt'

characters_columns = ['wikiID', 'movieID', 'releaseDate', 'charactName', 'birth', 'gender', 'height', 'ethnicity',
                      'name_actor', 'age', 'charActID', 'charID', 'actID']
movies_columns = ['wikiID', 'movieID', 'name_movie', 'releaseDate', 'revenue', 'runtime', 'languages',
                  'countries', 'genres']

characters = pd.read_table(CHARACTER_DATASET, header=None, names=characters_columns)
movies = pd.read_table(MOVIE_DATASET, header=None, names=movies_columns)
```

```python
characters.head()
```

```python
movies.head()
```

```python
# Make deep copies to make sure not to change original data
characters_copy = characters.copy()
movies_copy = movies.copy()
```

```python
# Drop rows with age that are NaN or negative 
characters_age = characters_copy.dropna(subset=['age'])
characters_age = characters_age[characters_age.age>=0]
```

```python
sns.histplot(characters_age['age'], kde=True, stat='density', bins=50)
```

```python
# Populate the character dataset with infromations about the movies
character_movies = pd.merge(characters_age, movies_copy, on=['wikiID', 'movieID', 'releaseDate'] , how='left')
```

```python
# Remove all rows with NaN values for actor name and movie revenue
character_movies = character_movies.dropna(subset=['name_actor', 'revenue'])

# Remove all rows with values other than M or F as gender
character_movies = character_movies[~((character_movies.gender != 'M') & (character_movies.gender != 'F'))]

print(len(characters_copy))
print(len(character_movies))
print(len(character_movies)/len(characters_copy)*100)

character_movies.head()
```

```python
# Sort between actors that appear only once and actors that appear multiple times
groupped_actor_name = character_movies.groupby(['name_actor']).size().reset_index().rename(columns={0: 'count'})
uniques = groupped_actor_name[groupped_actor_name['count'] <= 1]

character_movies['multiple'] = (~character_movies.name_actor.isin(uniques.name_actor)).astype(int)

unique_actors = character_movies[character_movies.multiple == 0]
duplicate_actors = character_movies[character_movies.multiple == 1]
```

```python
sns.histplot(unique_actors['revenue'], kde=True, stat='density', bins=50,
             color='blue', label='Actors that appear only once', log_scale=True)
sns.histplot(duplicate_actors['revenue'], kde=True, stat='density', bins=50,
             color='orange', label='Actors played in multiple movies', log_scale=True)

plt.title('Revenue of movies')
plt.legend()
plt.show()
```

```python
# The p_value for the difference in mean revenue btw movies with reccurent acotrs and movie with unique actors
stats.ttest_ind(unique_actors.revenue, duplicate_actors.revenue)

# Here p_value<<0.05 but much more filtering and analysis needs to be done (matching) but if this holds after
# a deeper analysis -> "The actors that play in many movies are usually more successful (movie revenue is higher) "
```

```python
# Find the actors that have played in a movie while underage (<=18) and that played in multiple movies
young_actors = character_movies[character_movies['age'] <= 18].name_actor

character_movies['start_young'] = character_movies.name_actor.isin(young_actors).astype(int)
```

```python
# Plot the number of movies an actor (that played at least in 2 movies) appears in
actors_start_young = character_movies[(character_movies.start_young==1) & (character_movies.multiple==1)]
actors_start_adult = character_movies[(character_movies.start_young==0) & (character_movies.multiple==1)]

nb_movies_start_young = actors_start_young.groupby(['name_actor']).wikiID.count()
nb_movies_start_adult = actors_start_adult.groupby(['name_actor']).wikiID.count()

print(f'The mean number of movies played by actors that started before 18 \
(that that played in at least two movies) is: {nb_movies_start_young.mean(): .2f}')
print(f'The mean number of movies played by actors that started after 18 \
(that that played in at least two movies) is: {nb_movies_start_adult.mean(): .2f}')

sns.histplot(nb_movies_start_young, kde=True, stat='density', bins=15,
             color='blue', label='Actors that started young')
sns.histplot(nb_movies_start_adult, kde=True, stat='density', bins=15,
             color='orange', label='Actors that started as adults')

plt.title('Number of movies played by actors')
plt.legend()
plt.show()
```

```python
# p_value for difference in mean -> same conclusion as before but p_value>0.05
stats.ttest_ind(nb_movies_start_young, nb_movies_start_adult)
```

```python
# Plot the distribution of movie revenue depending on wheter an actor from the movie started young
sns.histplot(actors_start_adult.revenue, kde=True, stat='density', bins=50,
             color='blue', label='Actors started as adults', log_scale=True)
sns.histplot(actors_start_young.revenue, kde=True, stat='density', bins=50,
             color='orange', label='Actors that started young', log_scale=True)

plt.title('Movie revenue')
plt.legend()
plt.show()
```

```python
# p_value for difference in mean -> same conclusion as before
stats.ttest_ind(actors_start_adult.revenue, actors_start_young.revenue)
```

```python
# Let's have a look at the distribution of age in our dataset

# First remove all duplicates (keep last to have actors at the end of their career)
character_movies_no_duplic = character_movies.copy()
character_movies_no_duplic.drop_duplicates(subset=['name_actor'], keep='last')

actors_no_duplic_start_young = character_movies_no_duplic[character_movies_no_duplic.start_young ==1]
actors_no_duplic_start_adult = character_movies_no_duplic[character_movies_no_duplic.start_young ==0]

fig = sns.boxplot(data=[actors_no_duplic_start_young['age'],
                  actors_no_duplic_start_adult['age']])

fig.set_title('Age')
fig.set_xticklabels(['young', 'adult'])
plt.ylabel('age')

plt.tight_layout()
```

```python
group = character_movies_no_duplic.groupby(['start_young', 'gender'])['gender'].count()
group = group.unstack()
group = group.div(group.sum(axis=1), axis=0)

pl = group.plot(kind='bar', figsize=[7,5], rot=0)
pl.set_title('gender')
pl.set_ylabel('participants')
pl.set_xlabel('Start young')
plt.show()
```

```python
# logistic regression model

# Standarize continuous variables of our model
character_movies_no_duplic['age_strd'] = (character_movies_no_duplic['age'] - character_movies_no_duplic['age'].mean())/character_movies_no_duplic['age'].std()

# Define the model
mod = smf.logit(formula='start_young ~  age_strd + C(gender)', data=character_movies_no_duplic)

res = mod.fit()

# Extract the estimated propensity scores
character_movies_no_duplic['Propensity_score'] = res.predict()
```

```python
# Balance the dataset via matching

def get_similarity(propensity_score1, propensity_score2):
    '''Calculate similarity for instances with given propensity scores'''
    return 1-np.abs(propensity_score1-propensity_score2)
```

```python
# Separate the treatment and control groups
actors_start_young = character_movies_no_duplic[character_movies_no_duplic.start_young == 1]
actors_start_adult = character_movies_no_duplic[character_movies_no_duplic.start_young == 0]
```

```python
# Runtime error  - dataset too long - 



# Create an empty undirected graph
G = nx.Graph()

# Loop through all the pairs of instances
for actors_start_adult_id, actors_start_adult_row in actors_start_adult.iterrows():
    for actors_start_young_id, actors_start_young_row in actors_start_young.iterrows():

        # Calculate the similarity q
        similarity = get_similarity(actors_start_adult_row['Propensity_score'],
                                    actors_start_young_row['Propensity_score'])

        # Add an edge between the two instances weighted by the similarity between them
        G.add_weighted_edges_from([(actors_start_adult_id, actors_start_young_id, similarity)])

# Generate and return the maximum weight matching on the generated graph
matching = nx.max_weight_matching(G)
```

```python
matched = [i[0] for i in list(matching)] + [i[1] for i in list(matching)]
```

```python
balanced_df = character_movies_no_duplic.iloc[matched]
```

```python
# Let's have a look at the new data distribution
actors_start_young_balanced = balanced_df[balanced_df.start_young == 1]
actors_start_adult_balanced = balanced_df[balanced_df.start_young == 0]

fig = sns.boxplot(data=[actors_start_young_balanced['age'],
                  actors_start_adult_balanced['age']])

fig.set_title('Age')
fig.set_xticklabels(['young', 'adult'])
plt.ylabel('age')

plt.tight_layout()
```

```python
group = balanced_df.groupby(['start_young', 'gender'])['gender'].count()
group = group.unstack()
group = group.div(group.sum(axis=1), axis=0)

pl = group.plot(kind='bar', figsize=[7,5], rot=0)
pl.set_title('gender')
pl.set_ylabel('participants')
pl.set_xlabel('Start young')
plt.show()
```
